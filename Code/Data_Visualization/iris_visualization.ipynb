```python
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn import datasets
import pandas as pd
```

```python
datas = sns.load_dataset('iris')
datas
```

```python
sepals_datas = datas[['sepal_length','sepal_width']]
petal_datas = datas[['petal_length','petal_width']] 
# sepals_datas = pd.DataFrame({"length":sepals_length,"width":sepals_width})
# petal_datas = pd.DataFrame({"length":petal_length,"width":petal_width})
```

```python

plt.figure(figsize=(10,10))
sns.scatterplot(x="sepal_length",y="sepal_width",data = sepals_datas,legend="full")
# sns.set(style='darkgrid',)
sns.scatterplot(x="petal_length",y="petal_width",data = petal_datas,legend="full")
plt.show()
```

```python
sns.set(style='darkgrid',)
sns.jointplot(data=sepals_datas, x="sepal_length", y="sepal_width")
plt.show()
```
